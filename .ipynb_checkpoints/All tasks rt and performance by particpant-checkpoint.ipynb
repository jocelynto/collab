```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
participant_codes = (1,2,3)
stroop_performance = {'correct': (105, 101, 101), 'incorrect': (3,5,5), 'missed': (0,2,2)}
clock_performance = {'correct': (66,70,62), 'incorrect': (11,10,6), 'missed': (3,0,12)}
nback_2_performance = {'correct': (52,54,60), 'incorrect': (25,26,20), 'missed': (3,0,0)}
nback_0_performance = {'correct': (61,62,49), 'incorrect': (16,18,29), 'missed': (3,0,2)}

# stroop
x_ticks = np.arange(len(participant_codes))
width = 0.25
multiplier = 0

fig1, ax1 = plt.subplots(layout='constrained')
for attribute, measurement in stroop_performance.items():
    offset = width * multiplier
    rects = ax1.bar(x_ticks + offset, measurement, width, label=attribute)
    ax1.bar_label(rects, padding=3)
    multiplier += 1

ax1.set_title('Stroop performance by participant code')
ax1.set_ylabel('performance')
ax1.set_xticks(x_ticks + width, participant_codes)
ax1.set_xlabel('participant code')
ax1.set_ylim(0, 120)
plt.legend()
fig1.savefig("Task checks/Stroop performance.png")

# clock
x_ticks = np.arange(len(participant_codes))
width = 0.25
multiplier = 0

fig2, ax2 = plt.subplots(layout='constrained')
for attribute, measurement in clock_performance.items():
    offset = width * multiplier
    rects = ax2.bar(x_ticks + offset, measurement, width, label=attribute)
    ax2.bar_label(rects, padding=3)
    multiplier += 1

ax2.set_title('Clock performance by participant code')
ax2.set_ylabel('performance')
ax2.set_xticks(x_ticks + width, participant_codes)
ax2.set_xlabel('participant code')
ax2.set_ylim(0, 80)
plt.legend()
fig2.savefig("Task checks/Clock performance.png")

# 2-back
x_ticks = np.arange(len(participant_codes))
width = 0.25
multiplier = 0

fig3, ax3 = plt.subplots(layout='constrained')
for attribute, measurement in nback_2_performance.items():
    offset = width * multiplier
    rects = ax3.bar(x_ticks + offset, measurement, width, label=attribute)
    ax3.bar_label(rects, padding=3)
    multiplier += 1

ax3.set_title('2-back performance by participant code')
ax3.set_ylabel('performance')
ax3.set_xticks(x_ticks + width, participant_codes)
ax3.set_xlabel('participant code')
ax3.set_ylim(0, 80)
plt.legend()
fig3.savefig("Task checks/2-back performance.png")

#0-back
fig4, ax4 = plt.subplots(layout='constrained')
for attribute, measurement in nback_0_performance.items():
    offset = width * multiplier
    rects = ax4.bar(x_ticks + offset, measurement, width, label=attribute)
    ax4.bar_label(rects, padding=3)
    multiplier += 1

ax4.set_title('0-back performance by participant code')
ax4.set_ylabel('performance')
ax4.set_xticks(x_ticks + width, participant_codes)
ax4.set_xlabel('participant code')
ax4.set_ylim(0, 80)
plt.legend()
fig4.savefig("Task checks/0-back performance.png")
```

```python
stroop_rt = [1.037772191, 0.790010452, 0.885216961]
clock_rt = [1.387421816, 1.42293723, 1.640260178]
nback_2_rt = [0.969171688, 0.733829582, 1.179979216]
nback_0_rt = [1.014790182, 0.782484232, 0.894755258]

# stroop
fig5 = plt.figure()
ax5 = fig5.add_subplot(1,1,1)
ax5.axhline(2,color='r', label='max allowed response time')
ax5.bar([1,2,3],stroop_rt,color='g')
ax5.set_title('Stroop task reaction time by participant code')
ax5.set_xlabel('participant code')
ax5.set_xticks([1,2,3])
ax5.set_ylabel('reaction time (s)')
plt.legend(loc=(0.5,0.85))
fig5.savefig('Task checks/stroop_rt.png')

# clock
fig6 = plt.figure()
ax6 = fig6.add_subplot(1,1,1)
ax6.axhline(2.5,color='r', label='max allowed response time')
ax6.bar([1,2,3],clock_rt,color='g')
ax6.set_title('Clock task reaction time by participant code')
ax6.set_xlabel('participant code')
ax6.set_xticks([1,2,3])
ax6.set_ylabel('reaction time (s)')
plt.legend(loc=(0.5,0.85))
fig6.savefig('Task checks/clock_rt.png')

# 2-back
fig7 = plt.figure()
ax7 = fig7.add_subplot(1,1,1)
ax7.axhline(2,color='r', label='max allowed response time')
ax7.bar([1,2,3],nback_2_rt,color='g')
ax7.set_title('2-back task reaction time by participant code')
ax7.set_xlabel('participant code')
ax7.set_xticks([1,2,3])
ax7.set_ylabel('reaction time (s)')
plt.legend(loc=(0.5,0.85))
fig7.savefig('Task checks/2_back_rt.png')

# 2-back
fig8 = plt.figure()
ax8 = fig8.add_subplot(1,1,1)
ax8.axhline(2,color='r', label='max allowed response time')
ax8.bar([1,2,3],nback_0_rt,color='g')
ax8.set_title('0-back task reaction time by participant code')
ax8.set_xlabel('participant code')
ax8.set_xticks([1,2,3])
ax8.set_ylabel('reaction time (s)')
plt.legend(loc=(0.5,0.85))
fig8.savefig('Task checks/0_back_rt.png')
```
